```python
import pandas as pd
import os
```

```python
folder="/home/durga/Desktop/DURGA/E2S1/DS/Course/NIFTY-Stock-Dashboard/dataset"
files=os.listdir(folder)
print("total files",len(files))
print("files",files)
```

```python
df=pd.read_csv(os.path.join(folder,files[0]))
df.head()
```

```python
df_reliance = df[df["Symbol"]=="RELIANCE"].copy()
```

```python
df_reliance
```

```python
df_reliance.isnull().sum()
```

```python
df_reliance.info()
```

```python
df_reliance.isnull().sum()
```

```python
df_reliance["Date"]=pd.to_datetime(df_reliance["Date"])
```

```python
df_reliance.info()
```

```python
df_reliance.fillna(df.mean(numeric_only=True),inplace=True)
```

```python
df_reliance["target"]=df_reliance["Close"].shift(-1)
```

```python
df_reliance
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="mean")
df_reliance["target"]=imputer.fit_transform(df_reliance[["target"]])
df_reliance
```

```python

df_reliance["Date"]=df_reliance["Date"].sort_values()
df_reliance
```

```python
#price change in that dte
df_reliance["Price Chane"]=df_reliance["Close"]-df_reliance["Open"]
```

```python
#daily rerturns
df_reliance["Daily Returns"]=((df_reliance["Close"]-df_reliance["Open"])/df_reliance["Open"])*100
#high and low differ
df_reliance["High Low Difference"]=df_reliance["High"]-df_reliance["Low"]
```

```python
df_reliance
```

```python
X = df_reliance[
    [
        "Prev Close",
        "Open",
        "High",
        "Low",
        "VWAP",
        "Volume",
        "Trades",
        "Deliverable Volume",
        "%Deliverble",
        "Price Chane",
        "High Low Difference",
        "Daily Returns"    ]
]
```

```python
X.shape
```

```python
y = df_reliance["target"]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test)
X_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
mae = mean_absolute_error(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)

rmse = mse ** 0.5

r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error :", mae)
print("Mean Squared Error :", mse)
print("Root Mean Squared Error :", rmse)
print("R2 Score :", r2)
```

```python
import pandas as pd

results = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
})

results.head(10)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)

dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(random_state=42)

gb.fit(X_train, y_train)

y_pred_gb = gb.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(model_name, y_test, y_pred):
    print(f"\n{model_name}")
    print("MAE :", mean_absolute_error(y_test, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
    print("R2  :", r2_score(y_test, y_pred))
```

```python
evaluate_model("Linear Regression", y_test, y_pred)
evaluate_model("Decision Tree", y_test, y_pred_dt)
evaluate_model("Random Forest", y_test, y_pred_rf)
evaluate_model("Gradient Boosting", y_test, y_pred_gb)
```

Purpose:

To compare the performance of different regression algorithms for predicting the next day's closing price of RELIANCE.

Observation:

Linear Regression achieved the highest R² score (0.9943) and the lowest MAE and RMSE among all tested models. Gradient Boosting and Random Forest also performed well, while Decision Tree produced comparatively larger prediction errors.

Conclusion:

Linear Regression was selected as the final model because it provided the best overall predictive performance on the test dataset.
